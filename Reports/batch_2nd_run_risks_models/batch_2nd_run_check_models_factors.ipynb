```python
def read_model_reports(conn, model_name, timestamp_min='2016-08-12 00:00:00', timestamp_max='2016-08-28 00:00:00'):
    sqlcmd = """
    select * from model.reports
    where model_name='%s'
    timestamp>'{ts_min}' and timestamp<'{ts_max}'
    """.format(ts_min=timestamp_min, ts_max=timestamp_max)
    df = pd.read_sql_query(sqlcmd, conn)
    return(df)
```
